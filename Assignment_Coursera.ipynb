```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

response = requests.get("https://stockanalysis.com/stocks/tsla/revenue/")
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print("Done")
else :
    print("Error to fetch data")

table = soup.find("table")

headers = [th.text.strip() for th in table.find_all("th")]

rows = []

for tr in table.find_all("tr")[1:]:  # Skip the header row
    cells = [td.text.strip() for td in tr.find_all("td")]
    rows.append(cells)


df = pd.DataFrame(rows, columns=headers)

df = df.drop(["Change","Growth"], axis = 1)
df.columns = ["Fiscal_Year_End","Revenue"]
print(df.head(5))
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def make_graph():
    df = yf.download("TSLA", period="10d", interval="1d")
    df = df.sort_index(ascending=False)
    df.reset_index(inplace=True)
    
    df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
    
    plt.figure(figsize=(12, 4)) 
    
    plt.plot(df["Date"], df["Close"], color="green")
    
    plt.ylabel("Closing Price (in $)")
    plt.title("Tesla Share Price Over Last 30 Days")
    plt.show()

make_graph()
```
